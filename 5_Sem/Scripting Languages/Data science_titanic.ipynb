```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
titanic_df=pd.read_csv('train.csv')
titanic_df.head()
```

```python
#Converting survived columns to strings
titanic_df["Survived"]=titanic_df['Survived'].map({0:"Died",1:"Survived"})
titanic_df.head()
```

```python
print("Data transformation after dropping columns")
titanic_df.drop(['Parch','Name','Ticket','PassengerId'],axis=1,inplace=True) #First way of dropping columns
titanic_df.head()
```

```python
titanic_df=titanic_df.drop(['SibSp','Fare'],axis=1)  #Second way of dropping columns
titanic_df.head()
```

```python
titanic_df['Pclass']=titanic_df['Pclass'].map({1:'Luxury Class',2:'Economy Class',3:'Lower Class'})
titanic_df.head()
```

```python
titanic_df['Embarked']=titanic_df['Embarked'].fillna("S")
# print(titanic_df.head())
titanic_df.head()
```

```python
titanic_df['Embarked']=titanic_df['Embarked'].map({'C':'Cherbourg','S':'Southampton','Q':'Queenstown'})
titanic_df.head()
```

```python
print('Data Visualizations')
print('1.Visualizing Survival rate based on Passenger Sitting class')
ax=sns.countplot(x='Pclass',hue='Survived',data=titanic_df,palette='Set1')
ax.set(xlabel="Passenger classes",ylabel="Survival rate",title="Survival rate VS Passenger sitting class")
plt.show()
```

```python
print('2.Visualizing Survival rate based on Gender')
ax=sns.countplot(x='Sex',hue='Survived',data=titanic_df,palette='Set2')
ax.set(xlabel="Gender",ylabel="Survival rate",title="Survival rate VS Gender")
plt.show()
```

```python
print('3.Visualizing Survival rate based on Passenger Age group')
titanic_df['Age']=titanic_df['Age'].fillna(20)
interval=(0,18,35,60,120)
categories=['Children','Teens','Adult','Old']
titanic_df['Age_cats']=pd.cut(titanic_df.Age,interval,labels=categories)
ax=sns.countplot(x='Age_cats',hue='Survived',data=titanic_df,palette='Set3')
ax.set(xlabel="Age groups",ylabel="Survival rate",title="Survival rate VS Age groups")
plt.show()
```
